## Análise 3

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
plot = sns.catplot(y='price', x='fuelType', hue='gearbox', kind='bar', data=df, palette='husl', height=6)
plot.ax.xaxis.set_label_text('Tipo de Combustível')
plot.ax.yaxis.set_label_text('Preço Médio')
plot.ax.set_title('Preço médio dos veículos por tipo de combustível e caixa de câmbio')
```

```python
# Salvando o plot
plot.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

```python
# Opção 2 de gerar os dados utilizando barplot do pacote seaborn
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(y=df.price,x=df.fuelType, hue=df.gearbox, palette='husl')
ax.set_ylabel('Preço Médio')
ax.set_xlabel('Tipo de Combustível')
ax.set_title('Preço médio dos veículos por tipo de combustível e caixa de câmbio')
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price-v2.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(y=df.powerPS,x=df.vehicleType, hue=df.gearbox, palette='husl')
ax.set_ylabel('Potência Média')
ax.set_xlabel('Tipo de Veículo')
ax.set_title('Potência média dos veículos por tipo de veículo e caixa de câmbio')
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```
