```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
sns.set_style('whitegrid') # Adiciona um estilo ao gráfico
fig, ax = plt.subplots(figsize=(12,10))

sns.countplot(y=df.brand, palette='Reds_r')
ax.set_ylabel('Marca', fontsize=20)
ax.set_xlabel('Número de Veículos', fontsize=20)
ax.set_title('Veículos por Marca', fontsize=20)
```

```python
# Salvando o plot
fig.savefig(("plots/Analise2/brand-vehicleCount.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
#help(sns.barplot)
```

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(8,6))
#hue -> Determina qual coluna será usada para separar/colorir os barplots
sns.barplot(y=df.price, x=df.vehicleType, hue=df.gearbox, palette=['#ff1112','#00ff00','#0000ff']) 
ax.set_xlabel('Tipo de Veículo')
ax.set_ylabel('Preço médio')
ax.set_title('Preço médio dos veículos por tipo de veículos e tipo de caixa de câmbio')
```

```python
# Salvando o plot
fig.savefig("plots/Analise2/vehicletype-gearbox-price.png")
```
