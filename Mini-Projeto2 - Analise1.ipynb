## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```python
df.isnull().sum()
# variável fuelType possui valores NULLS
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
fig, ax = plt.subplots(figsize=(8,6))

sns.distplot(df['yearOfRegistration'], kde=True) # kde = Densidade Kernel gaussian
ax.set_ylabel('Densidade (KDE)', fontsize= 15)
ax.set_xlabel('Ano de Registro', fontsize= 15)
ax.set_title('Distribuição de Veículos com base no Ano de Registro', fontsize= 15)
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
#sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(8,6))

sns.boxplot(df.vehicleType,  df.price)
#sns.boxplot(x="vehicleType", y="price", data=df)
ax.set_ylabel('Range de Preço', fontsize=15)
ax.set_xlabel('Tipo de Veículo', fontsize=15)
ax.text(x=5.25,y=27000,s='Análise de Outliers',color='r',fontsize=18) # adiciona texto dentro do plot
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/price-vehicleType-boxplot.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
ax.patches
```

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
sns.set_style("white")
fig, ax = plt.subplots(figsize=(9,6))
sns.countplot(df.vehicleType, palette="BuPu")
ax.set_ylabel('Total Veículos para venda', fontsize=15)
ax.set_xlabel('Tipo de Veículo', fontsize=15)
ax.set_title('Contagem total de veículos à venda conforme o tipo de veículo', fontsize=15)

# Cada patches representa uma barra criada no gráfico.

# Adiciona uma legenda sobre cada barra criado
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+500))
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/count-vehicleType.png")
```
